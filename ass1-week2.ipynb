problem 5:

```python
def python_functionalize(y,start,end,interval):
  def plotchart(x,y,x_label,y_label,title):
    plt.plot(x, y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()
  x = [i for i in np.arange(start,end + interval,interval)]
  x = np.array(x)
  dx = x[1:] - x[:-1]
  y = [y(i) for i in x]
  y = np.array(y)
  dy = y[1:] - y[:-1]
  gradient = [(dy[i]/dx[i]) for i in range(0,len(dx))]
  plotchart(x,y,"x - axis","y - axis","Linear chart")
  plotchart(x[1:],gradient,"x value","gradient value","Gradient chart")
  return  np.concatenate(([x.reshape(len(x),1)],[y.reshape(len(y),1)])),np.array(gradient)
def function_a(x):
  return x**2
def function_b(x):
  return 2*x**2 + 2**x
def function_c(x):
  return np.sin(x**1/2)
rs1= python_functionalize(function_a,-50,50,0.1)
rs2= python_functionalize(function_b,-50,50,0.1)
rs3= python_functionalize(function_c,0,50,0.1)
```
