Problem 1 : Creating random numbers

```python
import numpy as np
np.random.seed(0)
mean = (-3,0)
cov = [[1, 0.8], [0.8, 1]]
rs = np.random.multivariate_normal(mean, cov, 500) 
```

Problem 2: Visualization by scatter plot

```python
import matplotlib.pyplot as plt
x = []
y = []
for i in rs:
  x.append(i[0])
  y.append(i[1])
plt.title("Random number with seed 0")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y,color = 'red')
plt.show()
```

Problem 3: Visualization by histogram

```python
plt.title("Histogram of x")
plt.hist(x)
plt.xlim(-6, 0)  
plt.show()
plt.title("Histogram of y")
plt.hist(y,color = "green")
plt.xlim(-4, 4)  
plt.show()
```

Problem 4: Addition of data

```python
np.random.seed(200)
mean2 = (0,-3)
rs2 = np.random.multivariate_normal(mean2, cov, 500) 
x2 = []
y2 = []
for i in rs2:
  x2.append(i[0])
  y2.append(i[1])
plt.title("Random number ")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y,color = 'red')
plt.scatter(x2, y2,color = 'blue')
plt.show()
```

Problem 5: Data combination

```python
combined_array = np.concatenate((rs,rs2))
combined_array.reshape(1000,2)
print(combined_array)
```

Problem 6: Labeling

```python
#take out the first element and increase the size of the array and increase the dimension
new_rs = combined_array[0]
new_rs = np.concatenate((new_rs,np.array([0])))
new_rs = [new_rs]
for k in combined_array[1:500]:
  k = np.concatenate((k,np.array([0])))
  new_rs = np.concatenate((new_rs,[k]))

for k in combined_array[500:]:
  k = np.concatenate((k,np.array([1])))
  new_rs = np.concatenate((new_rs,[k]))
print(new_rs)
```
